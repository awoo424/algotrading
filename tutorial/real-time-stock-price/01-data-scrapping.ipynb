````python
# import libraries
from bs4 import BeautifulSoup
import requests_html
import pandas as pd
import time
import re
````

### 1.1 Scrapping real-time price for '0001.HK'

````python
# example
url = 'https://in.finance.yahoo.com/quote/{0}{1}'.format("0001", ".hk")

# HTML parser
session = requests_html.HTMLSession()
# make a GET request to the url 
r = session.get(url)

# create an instance of the BeautifulSoup class to parse the HTML document
soup = BeautifulSoup(r.content, 'lxml')

# inspect content
print(soup.prettify())
````

### 1.2 Scrapping data for a list of stock tickers

full list of symbols can be downloaded here: http://eoddata.com/stocklist/HKEX.htm

````python
# read HKEx list of tickers file

# converters option for keeping leading zeroes in ticker symbol
df = pd.read_csv('../../tickers/hkex_tickers.csv', converters={'symbol': lambda x: str(x)})

# inspect dataframe
df.head()
````

````python
# create list of tickers
tickers = df['symbol']

# traverse tickers list
for ticker in tickers:
    url = 'https://in.finance.yahoo.com/quote/{0}{1}'.format(ticker, ".hk")

    session = requests_html.HTMLSession()
    r = session.get(url)
    soup = BeautifulSoup(r.content, 'lxml')

    try:
        price = soup.select('table td')[5].text.split(' ')[0]
        price = float(price)
    except IndexError as e:
        price = None

    time.sleep(1)
    
    print("%s%s: %f" %(ticker, ".HK", price) )
    break
    
print('Completed')
````
